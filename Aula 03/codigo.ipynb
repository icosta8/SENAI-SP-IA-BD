```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0,10,10)
dados_y = dados_x * dados_x
```

```python
plt.figure(figsize=(10,8),facecolor='gray')
ax = plt.axes()
ax.set_facecolor('white')

plt.plot(dados_x, dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')
plt.xlabel('Eixo X', fontsize=16)
plt.ylabel('Eixo Y', fontsize=16)
plt.title('Gráfico simples com Matplotlib', fontsize=16)
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(12,12))
linhas = 2
colunas = 2

plt.subplot(linhas, colunas, 1)
plt.plot(dados_x, 'r--', label = 'Dados X')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico 1')

plt.subplot(linhas, colunas, 2)
plt.plot(dados_x, dados_y,'b--', label='Dados x e y')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico 2')

plt.subplot(linhas, colunas, 3)
plt.plot(dados_x* dados_y,'g--', label='Dados x^2')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico 3')

plt.subplot(linhas, colunas, 4)
plt.plot(dados_y,'c--', label='Dados y')
plt.legend(loc='lower right')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico 4')

plt.savefig('Graficos aula 03.png')
plt.show()
```

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,0.7,0.7])

rotulos = ['A', 'B', 'C', 'D', 'E','F', 'G', 'H', 'I', 'J']
grafico.bar(rotulos, dados_x, color='red')
grafico.set_xlabel('Rotulos')
grafico.set_ylabel('Quantidade')
```

```python
grafico.bar(rotulos, dados_x, color='red')
grafico.bar(rotulos, dados_y*0.1, color='blue')
figura
```

```python
grafico.bar(dados_x, dados_y/2, color=['red', 'blue'])
figura
```
