```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
df = pd.read_csv('C:\\Users\\dsadm\\Downloads\\IA_BD Isaque Costa\\Aula 03\\base_de_dados\\monitoramento_tempo.csv', sep=',')
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.figure(figsize=(15, 8))
plt.plot(df['data'], df['humidade'], label='Humidade', color='blue')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Umidade', fontsize=15)
plt.title('Umidade x Ano', fontsize=20)
plt.legend(loc='lower right')
plt.grid(True)

idx_max = df['humidade'].idxmax()
idx_max_ajustado = max(idx_max - 7000, 0)
x1 = df['data'][idx_max]
y1 = df['humidade'][idx_max]
x2 = df['data'][idx_max_ajustado]
y2 = y1 - 5
plt.annotate('Mximo', xy=(x1, y1), xytext=(x2, y2),
             fontsize=20, arrowprops=dict(facecolor='k'))

idx_min = df['humidade'].idxmin()
idx_min_ajustado = max(idx_min - 7000, 0)
x1 = df['data'][idx_min]
y1 = df['humidade'][idx_min]
x2 = df['data'][idx_min_ajustado]
y2 = y1 + 5
plt.annotate('Mnimo', xy=(x1, y1), xytext=(x2, y2),
             fontsize=20, arrowprops=dict(facecolor='k'))

plt.axhline(y1, color='k', linestyle='--')
plt.axhline(y1 - 10, color='k', linestyle='--')

plt.show()
```
