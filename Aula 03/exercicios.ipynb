```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
df = pd.read_csv('C:\\Users\\dsadm\\Downloads\\IA_BD Isaque Costa\\Aula 03\\base_de_dados\\monitoramento_tempo.csv', sep=',')
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'],df['humidade'], label='Umidade')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Umidade', fontsize=15)
plt.title('Umidade x Ano')
plt.legend(loc='lower right')
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'],df['velocidade do vento'], label='Velocidade do vento')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Velocidade do vento', fontsize=15)
plt.title('Velocidade do vento x Ano')
plt.legend(loc='lower right')
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'],df['pressão'], label='Pressão')
plt.xlabel('Data', fontsize=15)
plt.ylabel('Pressão', fontsize=15)
plt.title('Pressão x Ano')
plt.legend(loc='lower right')
```
